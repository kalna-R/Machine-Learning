```python
# draw a line in a diagram from (0,6) to (6,250)

xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# plot with 'x'
ypoints = np.array([2,5,55,8])

plt.plot(ypoints, marker= 'X')
plt.show()
```

```python
# plot data

ypoints = np.array([2,5,55,8])

# plot with 'O' and dotted line
plt.plot(ypoints, marker= 'o', linestyle = 'dotted')

# label axis
plt.xlabel("X axis")
plt.ylabel("Y axis")

# plt title
plt.title('All data shown here')

# show grid
plt.grid()

plt.show()
```
